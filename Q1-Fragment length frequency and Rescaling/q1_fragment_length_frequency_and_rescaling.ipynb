#### Rescaling

Often, you will received two sets of data with differnt emperical distribution.
The goal is to rescale one (query) to the other (reference). In this part, you
will write codes to achieve this. You are given a reference distribution
(`reference.hist`) and a query bed file (query.bed.gz). You have to rescale
(subsample the data) so that when you draw the normalized frequency, it should
almost match the reference. 

[reference.hist](https://figshare.com/ndownloader/files/53306810?private_link=727f8d920a1b8415f09a)

```python
# Let's import the necessary libraries

import gzip
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np
import pandas as pd
import random
```

```python
# Let's now extract the fragment lengths from the BED file, and store them in a list
# The BED file is gzipped, so we will use gzip to read it

fragment_lengths = []

with gzip.open('./data/query.bed.gz', 'rt') as f:
    for line in f:
        if line.strip() == '':
            continue
        fields = line.strip().split('\t')
        start = int(fields[1])
        end = int(fields[2])
        length = end - start
        fragment_lengths.append(length)
```

```python
# Count frequencies of each fragment length
# We will use Counter to count the occurrences of each length
length_counts = Counter(fragment_lengths)


# Normalize the frequencies
total_fragments = sum(length_counts.values())
normalized_freq = {k: v / total_fragments for k, v in length_counts.items()}
```

```python
# Sort by length for smooth plotting
sorted_lengths = sorted(normalized_freq.keys())
sorted_freqs = [normalized_freq[l] for l in sorted_lengths]


# Plotting the fragment length distribution
plt.figure(figsize=(10,5))
plt.plot(sorted_lengths, sorted_freqs, label='Query')
plt.xlabel("Fragment Length (bp)")
plt.ylabel("Normalized Frequency [A.U.]")
plt.title("Fragment Length Distribution (Query)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Let's now read the .hist file for the reference genome, which contains the expected fragment length distribution
# We will use pandas to read the file and create a dictionary for the reference distribution

reference_df = pd.read_csv('./data/reference.hist', sep='\t', header=None, names=['length', 'freq'])
reference_dict = dict(zip(reference_df['length'], reference_df['freq']))
```

```python
ref_lengths = sorted(reference_dict.keys())
ref_freqs = [reference_dict[l] for l in ref_lengths]

plt.figure(figsize=(10,5))
plt.plot(sorted_lengths, sorted_freqs, label='Query', alpha=0.7)
plt.plot(ref_lengths, ref_freqs, label='Reference', alpha=0.7)
plt.xlabel("Fragment Length (bp)")
plt.ylabel("Normalized Frequency [A.U.]")
plt.title("Query vs Reference Fragment Length Distribution")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Count original fragments by length
frag_by_length = {}
for l in fragment_lengths:
    frag_by_length.setdefault(l, []).append(l)


print("Total fragments in query:", total_fragments)

# Let's take about 30% of the total fragments from the reference distribution to rescale.
target_total = int(total_fragments * 0.3)

rescaled_fragments = []

for l, ref_freq in reference_dict.items():
    if l in frag_by_length:
        # Desired count based on reference proportion
        desired_count = int(ref_freq * target_total)

        current_pool = frag_by_length[l]
        sampled = random.sample(current_pool, min(len(current_pool), desired_count))
        rescaled_fragments.extend(sampled)

```

```python
# Let's count the rescaled fragments and normalize the frequencies

rescaled_counts = Counter(rescaled_fragments)
total_rescaled = sum(rescaled_counts.values())
rescaled_freq = {k: v / total_rescaled for k, v in rescaled_counts.items()}
```

```python
import matplotlib.pyplot as plt

# Convert dictionaries to sorted arrays for plotting
def dict_to_xy(d):
    x = sorted(d.keys())
    y = [d[k] for k in x]
    return x, y

query_x, query_y = dict_to_xy(normalized_freq)
ref_x, ref_y = dict_to_xy(reference_dict)
rescaled_x, rescaled_y = dict_to_xy(rescaled_freq)

plt.figure(figsize=(18, 8))
plt.plot(query_x, query_y, label=f'Query (n = {total_fragments:,})', color='magenta')
plt.plot(ref_x, ref_y, label='Reference', color='green')
plt.plot(rescaled_x, rescaled_y, 'x', label=f'Rep 0 (n = {total_rescaled:,})', color='deepskyblue', markersize=8, markeredgewidth=2)
plt.xlabel("Fragment Length (bp)", fontsize=24)
plt.ylabel("Normalized Frequency [A.U.]", fontsize=24)
plt.title("Rescaled Fragment Length Distribution", fontsize=24)
plt.xlim(0, 700)
plt.ylim(0, 0.025)
plt.legend(fontsize=14)  # Set font size for the legend
plt.tight_layout()
plt.savefig("./output_files/rescaled_prediction.png")
plt.show()
```
